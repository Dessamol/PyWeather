```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import os
import pprint

from config import api_key

from citipy import citipy

data_file = os.path.join("C:\\Users\\mas34\\Downloads\\output_data.csv")

lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    city = city.replace(" ", "%20")
    
    if city not in cities:

        cities.append(city)

len(cities)
```

```python

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key
```

```python
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

record = 1

print(f"Beginning Data Retrieval")
print(f"-------------------------------")

for city in cities:  
    
    try: 
        response = requests.get(f"{url}&q={city}").json() 
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"Processing Record {record} | {city_record}")
        print(f"{url}&q={city}")
        
        record= record + 1
        
        time.sleep(1.01)
    except:
                
        print("City not found. Skipping...")
        
    continue
```

```python
weatherpydict = {
    "City": city_name,
    "Cloudiness":cloudiness, 
    "Country":country,
    "Date":date, 
    "Humidity": humidity,
    "Lat":lat, 
    "Lng":lng, 
    "Max Temp": max_temp,
    "Wind Speed":wind_speed
}

weather_data = pd.DataFrame(weatherpydict)

weather_data.count()
```

```python
weather_data.to_csv("C:\\Users\\mas34\\Downloads\\output_data.csv")

weather_data.head()
```

```python
plt.scatter(weather_data["Lat"], weather_data["Max Temp"], marker=">", s=10)

plt.title("City Latitude vs. Max Temperature")
plt.ylabel("Max. Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

plt.savefig("Max_Temp_vs_Latitude.png")

plt.show()
```
